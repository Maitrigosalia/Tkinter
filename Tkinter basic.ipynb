```python
from tkinter import *
def Add():
    try:
        no=int(eno.get())
        num=int(enum.get())
        a=no+num
        lls.config(text=f"Sum = {a}")
    except:
        lls.config(text="Error")
def Sub():
    try:
        no=int(eno.get())
        num=int(enum.get())
        a=no-num
        lls.config(text=f"Substraction:{a}")
    except:
        lls.config(text="Error")
def Multiply():
    try:
        no=int(eno.get())
        num=int(enum.get())
        a=no*num
        lls.config(text=f"Multiplication:{a}")
    except:
        lls.config(text="Error")

def Division():
    try:
        no=int(eno.get())
        num=int(enum.get())
        a=no/num
        lls.config(text=f"Division:{a}")
    except:
        lls.config(text="Error")
root=Tk()
root.geometry("500x500")
root.title("Calculator")

no=Label(root,text="Enter Number",font=("Times",12,"bold"))
no.pack()
eno=Entry(root)
eno.pack()
num=Label(root,text="Enter Number",font=("Times",12,"bold"))
num.pack()
enum=Entry(root)
enum.pack()

b1=Button(text="Addition",command=Add)
b1.pack()

b2=Button(text="Substraction",command=Sub)
b2.pack()

b3=Button(text="Multiplication",command=Multiply)
b3.pack()

b4=Button(text="Division",command=Division)
b4.pack()

lls=Label(text="")
lls.pack()

root.mainloop()
```

```python
from tkinter import *
def exit():
    root.destroy()
def Farhenite():
    try:
        t=int(et.get())
        a=(9/5 * t) + 32.
        lls.config(text=f"Celsius to Farhenite:{a}")
    except:
        lls.config(text="Error")
def Celsius():
    try:
        t=int(et.get())
        a=(t - 32) * 5/9
        lls.config(text=f"Farhenite to Celsius:{a}")
    except:
        lls.config("Error")
root=Tk()
root.geometry("500x500")
root.title("Temperature Conversion")

t=Label(root,text="Enter Temperature",font=("Times",15,"bold"))
t.pack()
et=Entry(root)
et.pack()

b1=Button(text="Farhenite",command=Farhenite)
b1.pack()
b2=Button(text="Celsius",command=Celsius)
b2.pack()
b3=Button(text="Exit",command=exit)
b3.pack()
lls=Label(text="")
lls.pack()

root.mainloop()
```
